for the sake of extracting data from yahoo finance site we are going to use Pandas_datareader 
you find more informations here : https://pandas-datareader.readthedocs.io/en/latest/

```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import numpy as np
import warnings ## to delete long warnings 
warnings.simplefilter("ignore")
```

```python
## create the dictionary which will be used to extract the data for
companies_dict = {
    'Amazon' : 'AMZN',
    'Apple':'AAPL',
    'Walgreen':'WBA',
    'McDonalds':'MCD',
    'MasterCard':'MCD'
                   }
```

```python
data_source = 'yahoo'
start_date = '2010-10-10'
end_date = '2021-10-10'
df = data.DataReader(list(companies_dict.values()),data_source,start_date,end_date)
```

```python
## starting the split of the stock open and the stock close
stock_open = np.array(df['Open']).T ## stock open is numpy array of transpose of df['open']
stock_close = np.array(df['Close']).T ## stock close is numpy array of transpose of df['close']
```
